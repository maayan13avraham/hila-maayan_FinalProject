# e. Key Experiments, Evaluation, and Findings

**Experiments Conducted**
The experiments conducted in this project focused on verifying functional correctness, runtime efficiency, and basic retrieval quality, rather than on extensive model comparison or advanced parameter tuning.

The evaluation process consisted of the following main experiments.

**Functional Correctness Verification**

All implemented search endpoints (/search, /search_body, /search_title, /search_anchor) were tested using custom self-check scripts developed by the authors with the assistance of artificial intelligence.

These scripts were used to verify compliance with the required API interface and expected output format.

The tests verified:

*   Correct response structure
*   Valid HTTP responses
*   Proper query handling
*   Compliance with the required list-based output format

Evaluation method: Pass/fail based on the output of the self-check scripts.
Result: All endpoints passed the functional correctness tests successfully.

**Retrieval Quality Evaluation (Sanity Check)**

To evaluate retrieval quality, we used a custom AP@10 evaluation script that computes Average Precision@10 using pseudo-relevance judgments derived from Wikipedia titles and anchor text.

The evaluation was performed over a predefined set of sanity queries.

Evaluation method: Mean AP@10 over the evaluated queries.

Result:


*   Required minimum: AP@10 > 0.1
*   Mean AP@10  0.65

**Finding:** The engine significantly exceeds the minimum quality threshold. This demonstrates that a simple TF-IDF-based ranking over the document body, combined with binary matching on titles and anchor text, provides reasonable retrieval performance for a minimal search engine.

Runtime Performance Evaluation Using eval_queries.py

Runtime performance was evaluated using a dedicated query evaluation script (eval_queries.py).
This script reads a list of 30 predefined queries from a JSON file and sends them sequentially to the /search endpoint of the server.

For each query, the script measures:


*   Server-side response time (returned by the engine)
*   Client-side end-to-end response time
*   Precision@10, Recall@10, and AP@10

**Evaluation method:** Automatic measurement of response times and retrieval metrics for each query.

**Result:**
*   Mean client-side response time: 2.76 seconds
*   All queries completed successfully
*   No query exceeded the 35-second time limit
*   The slowest queries completed in under 5 seconds

**Finding:** The system consistently meets the runtime requirements. After initial server startup and warm-up, query execution times remain stable and well within the allowed limits.

**Key Findings**
*  A minimal IR system using TF-IDF scoring on the document body and binary matching on titles and anchors can achieve acceptable retrieval quality.
*   Custom evaluation scripts developed with the assistance of artificial intelligence are effective for validating correctness, retrieval quality, and runtime performance.
*   Loading index metadata and posting locations at server startup significantly improves query-time performance.
*   Query ambiguity remains the dominant cause of poor retrieval results and is not addressed in this minimal implementation.

Each version represents the addition of a new ranking signal, while keeping the system minimal and efficient.

The evaluated versions were:

  Ranking based solely on cosine similarity using TF-IDF on the article body.

- **Version 2  Body + Title**  
  TF-IDF body ranking combined with binary matching on article titles.

- **Version 3  Body + Title + Anchor (Final Version)**

The same set of sanity-check queries was used for all versions.

### Findings

The performance graph (**Figure f**) shows a clear improvement across versions.  
Adding title information significantly improves ranking quality, and incorporating anchor text provides additional gains.  

This demonstrates that combining multiple simple ranking signals leads to better retrieval quality, even without advanced models such as BM25 or PageRank.

  This reflects weaker top-10 relevance and is consistent with the qualitative findings

### Dominant Factor Behind Poor Results

The dominant factor behind the poorer performance on the query **"united states"** is
**query ambiguity**.

The query is very broad and can refer to many different aspects such as geography,
history, government, foreign policy, culture, or education.  
As a result, a large number of documents match the query terms, making it difficult for
a minimal, lexically-based ranking approach to prioritize a single canonical result.

---

### Possible Improvements

Several improvements could help address this limitation in future work:

- Incorporating **global importance signals** such as PageRank or page views
- Applying **query expansion** or user intent detection techniques
- Using more advanced ranking models such as **BM25** or semantic retrieval methods

These additions could help the engine better handle broad and ambiguous queries while
maintaining efficiency.
