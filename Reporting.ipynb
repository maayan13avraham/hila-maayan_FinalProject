# e. Key Experiments, Evaluation, and Findings Experiments Conducted

The experiments conducted in this project focused on verifying correctness, efficiency, and minimal retrieval quality, rather than on extensive model comparison or advanced parameter tuning.

The main experiments were:

1. Functional Correctness Verification

All implemented search endpoints (/search, /search_body, /search_title, /search_anchor) were tested using custom self-check scripts created by the authors with the assistance of artificial intelligence.
These scripts were designed to validate compliance with the required API interface.

The tests verified:

Correct response format

Valid HTTP responses

Proper query handling

Compliance with the expected output structure

Evaluation method:
Pass/fail based on the output of the self-check scripts.

Result:
All endpoints passed the functional correctness tests successfully.

2. Retrieval Quality Evaluation (AP@10 Sanity Check)

To evaluate retrieval quality, we used a custom AP@10 evaluation script, developed by the authors with the assistance of artificial intelligence.
The script computes Average Precision@10 using pseudo-qrels derived from Wikipedia titles and anchor text.

Evaluation method:
Mean AP@10 across a predefined set of sanity queries.

Result:

Mean AP@10  0.55

Required minimum: AP@10 > 0.1

Finding:
The engine significantly exceeds the minimum quality threshold, demonstrating that a simple TF-IDF-based body ranking combined with binary title and anchor matching provides reasonable retrieval performance.

3. Runtime Performance Measurement

Query execution time was measured using:

Timing instrumentation embedded in the server code

Timing information returned as part of each query response

The measurements included:

Total query time

Body processing time

Title processing time

Anchor processing time

Evaluation method:
Manual inspection of returned and logged timings.

Result:
Typical query execution time was well below one second, and no query exceeded the 35-second limit.

Finding:
Direct reading of posting lists from GCS, combined with lightweight scoring functions, is sufficient to meet the efficiency requirements of a minimal search engine.

Key Takeaways

A minimal IR system using TF-IDF on the document body and binary matching on titles and anchors can achieve acceptable retrieval quality.

Custom evaluation scripts created with the assistance of artificial intelligence are effective for validating correctness, quality, and performance.

Loading index metadata at server startup significantly improves query-time performance.

Query ambiguity remains the dominant factor behind poor retrieval results and is not addressed in this minimal implementation.
